arithmetico-geometric sequence

```python
import matplotlib.pyplot as plt
```

```python
def sum_of_ag_seq (n):
    result = 0
    for i in range(n):
        result = result + 1/3 * (2/3)**i *4*(i+1)
    return result
```

```python
sum_of_ag_seq (20)
```

```python
sum_set = []
for j in range(20):
    sum_set.append(sum_of_ag_seq(j))
plt.plot(sum_set)
plt.xlabel("Number of terms")
plt.ylabel("Sum")
plt.grid()
```
